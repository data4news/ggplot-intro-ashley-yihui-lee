### Dotplot (with jitter)

The jitter geom is a convenient shortcut for geom_point(position = "jitter"). It adds a small amount of random variation to the location of each point, and is a useful way of handling overplotting caused by discreteness in smaller datasets.

````python
%%R -w 1000 -h 200

ggplot(df) + 
    aes(x=pct_below_poverty, y=1) +
    geom_jitter()
````

## NOTES

In the line `aes(x=pct_below_poverty, y=1)`, the `y=1` is essentially fixing the `y` position of all your points at 1. This means all the data points are plotted at a constant `y` value of 1, and they will only vary along the `x` axis (which represents the `pct_below_poverty` variable). 

The `geom_jitter()` function adds a small random variation to the position of the points along the `y` axis to help prevent overplotting (i.e., when multiple points fall directly on top of each other). So, even though all your points are at `y=1`, jittering shifts them slightly up or down along the `y` axis in a random way to make each point visible.

Regarding the scatter, you're seeing points spread out because jitter is adding a random shift to each point's `y` value. This prevents the points from being stacked directly on top of each other (which would make them harder to see). The effect you're noticing where the points appear more scattered is due to this random jittering. 

In summary:
- The `y=1` means all points are initially at the same `y` coordinate (1).
- The `geom_jitter()` function applies a small, random shift to the `y` position to avoid overplotting and make the points more visible, which leads to the scattered look.

Install the package above and show the density as a beeswarm rather than just a jitter.


````python
%%R
library(ggplot2)
library(ggbeeswarm)

# Compare jitter and beeswarm
ggplot(df, aes(x = pct_below_poverty)) + 
  geom_jitter(width = 0.1, color = 'blue', alpha = 0.5) +   # Jitter plot
  ggtitle("Comparison: Jitter vs Beeswarm") +
  theme_minimal()

ggplot(df, aes(x = pct_below_poverty)) + 
  geom_beeswarm(color = 'red', alpha = 0.5) +   # Beeswarm plot
  ggtitle("Beeswarm Plot") +
  theme_minimal()

````
